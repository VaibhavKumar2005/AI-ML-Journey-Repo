```python
import matplotlib.pyplot as plt

years = [1880, 1900, 1920, 1940, 1960, 1980, 2000, 2020, 2023]
land_temp = [-0.12, -0.08, -0.05, 0.02, -0.02, 0.15, 0.45, 0.98, 1.15]
ocean_temp = [-0.22, -0.14, -0.10, -0.01, -0.08, 0.10, 0.35, 0.78, 0.90]
```

```python
# Line Chart: Land Temperature Only
plt.plot(years, land_temp, color='red')
plt.title('Global Land Temperature Anomalies (1880–2023)')
plt.xlabel('Year')
plt.ylabel('Temperature Anomaly (°C)')
plt.show()
```

```python
# Adding Grid and Marker to line chart
plt.figure(figsize=(10, 5))
plt.plot(years, land_temp, marker='o', color='red')
plt.title('Global Land Temperature Anomalies (1880–2023)')
plt.xlabel('Year')
plt.ylabel('Temperature Anomaly (°C)')
plt.grid(True)
plt.show()
```

```python
import pandas as pd

df = pd.read_excel('/content/Carbon_emission (1).xlsx')
df.head()
```

```python
# Set the x-axis to be the year
years = df['Year']

# Plot each country's data
plt.plot(years, df['USA'], label='USA', marker='o')
plt.plot(years, df['China'], label='China', marker='s')
plt.plot(years, df['India'], label='India', marker='^')

# Add title and labels
plt.title('CO2 Emissions by Country (2000–2020)',fontdict={"fontsize":15,"fontweight":"bold",'color':'red'})
plt.xlabel('Year')
plt.ylabel('CO2 Emissions (Million Metric Tons)')

# Add legend and grid
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2,figsize=(12, 6))

# First subplot
ax1.plot(years, df['USA'])
ax1.set_title('Carbin Emission USA')
plt.xlim(2000, 2025)
plt.ylim(0, 100)


# Second subplot
ax2.plot(years, df['India'])
ax2.set_title('Carbin Emission India')
plt.xlim(2000, 2025)
plt.ylim(0, 100)

plt.show()
```

```python
fig, ax = plt.subplots(2, 2) # Adjusted figsize for better visibility and space for titles/labels

# First subplot (top-left)
ax[0, 0].plot(years, df['USA'])
ax[0, 0].set_title('USA Carbon Emission')
ax[0, 0].set_xlabel('Year')
ax[0, 0].set_ylabel('CO2 Emissions')
ax[0, 0].set_xlim(2000, 2025)
ax[0, 0].set_ylim(0, 100)


# Second subplot (top-right)
ax[0, 1].plot(years, df['China'])
ax[0, 1].set_title('China Carbon Emission')
ax[0, 1].set_xlabel('Year')
ax[0, 1].set_ylabel('CO2 Emissions')
ax[0, 1].set_xlim(2000, 2025)
ax[0, 1].set_ylim(450, 600) # Adjusted y-limit based on data


# Third subplot (bottom-left)
ax[1, 0].plot(years, df['India'])
ax[1, 0].set_title('India Carbon Emission',fontsize=16)
ax[1, 0].set_xlabel('Year')
ax[1, 0].set_ylabel('CO2 Emissions')
ax[1, 0].set_xlim(2000, 2025)
ax[1, 0].set_ylim(30, 100) # Adjusted y-limit based on data

# Fourth subplot (bottom-right) - You can choose another column or leave it as is
ax[1, 1].plot(years, df['USA'],color="red" )
ax[1, 1].set_title('USA Carbon Emission (Repeat)')
ax[1, 1].set_xlabel('Year')
ax[1, 1].set_ylabel('CO2 Emissions')
ax[1, 1].set_xlim(2000, 2025)
ax[1, 1].set_ylim(0, 100)


fig.suptitle('Carbon Emissions by Country Over Time', fontsize=16)

plt.tight_layout() # Adjust layout to prevent overlapping titles/labels
plt.show()
```
