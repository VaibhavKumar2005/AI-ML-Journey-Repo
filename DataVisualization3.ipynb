```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
marks = [70, 60, 55, 68, 50]
subjects = ["Hindi", "English", " Science", "Maths", "GK"]
plt.pie(marks, labels = subjects, autopct = "%0.1f", explode = [0, 0.1, 0, 0, 0], shadow = True,
        colors = ["b", "g", "r", "violet", "gold"],
        startangle= 290,
        wedgeprops={"width" : .5},

        pctdistance = 0.8,
        textprops = {"color": "g", "fontweight" : "bold"})


plt.title("Marks")
plt.show()
```

```python
tips = sns.load_dataset("tips")
tips
ddaywisetotalbill = tips.groupby("day")["total_bill"].sum().reset_index()
ddaywisetotalbill
```

```python
from matplotlib.patches import Circle

plt.pie(x = ddaywisetotalbill ["total_bill"],
        labels = ddaywisetotalbill["day"], autopct = "0.2f")
centre_circle = Circle((0,0), 0.40, fc = "white")
fig = plt.gcf()
fig.gca().add_artist(centre_circle)
plt.title("Daywise Total Bill")
plt.show()
```
